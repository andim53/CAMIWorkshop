# 4. Add descriptors to the data ("featurization")

We are seeking to find relationships between the inputs (composition and crystal structure of a material) and outputs (elastic properties such as ``K_VRH``, and ``G_VRH``). To find such relations, we need to "featurize" the input data such that they are numbers that meaningfully represent the underlying physical quantity. For example, one "feature" or "descriptor" of the composition of a material such as ``Nb4CoSi`` would be the standard deviation of the Pauling electronegativity of the elements in the compound (weighted by stoichiometry). Compositions with high values of this quantity would be more ionic and quantities with lower values would tend towards covalent or ionic. A descriptor of the crystal structure might be the average coordination number of sites; higher coordination numbers indicate more bonds and therefore might indicate stiffer materials. With matminer, we can start generating hundreds of possible descriptors using the descriptor library that is available. Data mining techniques can help narrow down which descriptors are the most relevant to the target problem using the available output data as a guide.

The results of the model training and validation can produce "nan" (not a number). This suggests that there may be an issue with the model or the data being used. The loss values should ideally be numerical and decreasing over the epochs, indicating that the model is learning and improving its predictions. However, in this case, the loss values remain undefined throughout the training process, which indicates that the model is not able to learn from the data or make meaningful predictions. It is important to investigate the cause of this issue, such as checking the data quality, preprocessing steps, or the model architecture, to address the problem and improve the model's performance.

The model is compiled using the mean absolute error (MAE) as the loss function, which measures the average absolute difference between the predicted and actual values. The Adam optimizer is used to optimize the model's parameters during training.

MAE loss function follows this equation:


```
loss = mean(abs(y_true - y_pred))
```



Adam optimization is a stochastic gradient descent method that is based on adaptive estimation of first-order and second-order moments.

According to Kingma et al., 2014, the method is "computationally efficient, has little memory requirement, invariant to diagonal rescaling of gradients, and is well suited for problems that are large in terms of data/parameters".

## 5.3. Adding hidden layer into the model

Increasing the number of layers in a deep learning model can have both advantages and challenges. Adding more layers allows the model to learn more complex representations and capture intricate patterns in the data. This can potentially improve the model's performance and ability to extract meaningful features.

However, increasing the number of layers also introduces some challenges. Deeper models are generally more computationally expensive and require more training data to avoid overfitting. They may also be more prone to vanishing or exploding gradients, which can hinder the training process. Additionally, deeper models may be more difficult to optimize and require more careful hyperparameter tuning.

## 5.3. Iterating through number of neuron



```
neuron_mult = [1,2,3,4,5,6,7,8,9]

results = {}

def build_and_train_model(num_mult):

    input_layer = tf.keras.layers.Input(shape=(feat_col,))

    model = tf.keras.Sequential([

    history = model.fit(

    loss = history.history['val_loss'][-1]

    return loss

for num_mult in neuron_mult: # adding iteration
```

````python
neuron_mult = [1,2,3,4,5,6,7,8,9]

results = {}

def build_and_train_model(num_mult):

    input_layer = tf.keras.layers.Input(shape=(feat_col,))

    normalizer = tf.keras.layers.Normalization(axis=-1)
    normalizer.adapt(np.array(X))

    model = tf.keras.Sequential([
            input_layer,
            normalizer,
            tf.keras.layers.Dense(int(feat_col*num_mult), activation='relu'),
            tf.keras.layers.Dense(1)
        ])

    model.compile(loss='mean_absolute_error',
                  optimizer="adam")

    history = model.fit(
        np.array(X_train),
        np.array(y_train),
        epochs=10,
        verbose = 0,
        validation_split=0.2,
        callbacks=[tf.keras.callbacks.EarlyStopping(patience=3)],
        batch_size=32
    )

    loss = history.history['val_loss'][-1]

    return loss

for num_mult in neuron_mult: # adding iteration
    print(f"Number of neuron: {feat_col*num_mult}")
    loss = build_and_train_model(num_mult)
    print(f"Loss value: {loss}")
    print("-"*50)
    results[num_mult] = loss
````

## 5.4. Making the best model

````python
input_layer = tf.keras.layers.Input(shape=(feat_col,))

normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(np.array(X))

model = tf.keras.Sequential([
        input_layer,
        normalizer,
        tf.keras.layers.Dense(int(feat_col)*6, activation='relu'), # adding activation funciton
        tf.keras.layers.Dense(1)
    ])

model.compile(loss='mean_absolute_error',
              optimizer="adam")

history = model.fit(
    np.array(X_train),
    np.array(y_train),
    epochs=10,
    validation_split=0.2,
    callbacks=[tf.keras.callbacks.EarlyStopping(patience=3)],
    batch_size=32
)
````
